## **Matplotlib and seaborn**

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic data from seaborn library
df = sns.load_dataset('titanic')
```

```python
# scatter plot size or figure size
plt.figure(figsize=(8,4))

# change the scatter point colors
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data = df, x = 'age', y = 'fare', hue = 'sex', palette = palette)

# labels
xlab = 'Age'
ylab = 'Fare'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title('Scatter Plot of Age vs Fare')

# saving a plot 300dpi, 600dpi, 150dpi png
plt.savefig('/content/Plots/scatter_plot.png', dpi=300)

# saving a plot in vector file such as svg
plt.savefig('/content/Plots/scatter_plot.svg')
```

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1.5)
# scatter plot size or figure size
plt.figure(figsize=(8, 4)) # a4 page size 16:9

# change the scatter point colors
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', size='fare', palette=palette)
# labels
xlab = 'Age'
ylab = 'Fare'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title('Scatter Plot of Age vs Fare')
```

Interpretation:

1. We have Age of passengers at x-axis and fare at y-axis.

2. The scatter plot shows the relationship between age and fare of titanic passengers.

3. We further grouped the passengers by their genders such as male (Green points) and female (Red points)

4. Two male passengers gave the highest fare in titatnic and they might be travelling first class.

5. The oldest person at titanic is also a male having 80 years of age as shown in the scatter plot. (As shown in Figure. 1).

6. There doesn't seem to be a relationship between age and fare.

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1)
# box plot size or figure size
plt.figure(figsize=(12, 6)) # a4 page size 16:9

# change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'Blue'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)
# labels
xlab = 'Gender'
ylab = 'Age'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title('Boxen Plot of Age based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')
```

```python
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1)
# box plot size or figure size
plt.figure(figsize=(12, 6)) # a4 page size 16:9

# change the scatter point colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'Blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# labels
xlab = 'Gender'
ylab = 'Age'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title('Boxen Plot of Age based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=0)
```

```python
# scatter plot with plotly
import plotly.express as px
fig = px.scatter(df, x='age', y='fare', color='sex')
fig.show()
```

```python
# make scatter plot via plotly
import plotly.express as px
fig = px.scatter(df, x = 'age', y = 'fare', color = 'class')
fig.show()
```

```python
# bar plot using seaborn
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1, font='Times New Roman')
# bar plot size or figure size
plt.figure(figsize=(12, 6)) # a4 page size 16:9

# change the bar colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'Blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette, capsize=0.1)
# labels
xlab = 'Gender'
ylab = 'Age'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title('Bar Plot of Age based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=0)
```

```python
# violin plot using seaborn
# style of a plot
sns.set_theme(style='darkgrid', font_scale=1)
# violin plot size or figure size
plt.figure(figsize=(12, 6)) # a4 page size 16:9

# change the violin colors
palette = {'First': 'green', 'Second': 'red', 'Third': 'Blue'}
sns.violinplot(data=df, x='sex', y='age', hue='class', palette=palette)
# labels
xlab = 'Gender'
ylab = 'Age'
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title('Bar Plot of Age based on Gender and Class Name of Passengers')

# change legend position and title
plt.legend(loc='upper right', title='Class Name')

# change the x-ticks names
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=0)
```
